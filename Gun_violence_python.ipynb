```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
gv = ("GV.csv")

#Convert to DataFrames
gv_data = pd.read_csv(gv)

#Preview Rotten Tomatoes data
gv_data.head()
```

```python
new_gv = gv_data[["incident_id", "date", "state", "city_or_county",
                  "address","n_killed","n_injured","incident_url",
                  "participant_age","participant_age_group (Ron)",
                  "participant_gender","participant_status",
                  "participant_type","state_house_district",
                  "state_senate_district"]]
new_gv
```

```python
new_gv =new_gv.rename(columns={"participant_age_group (Ron)":"participant_age_group"})
```

```python
new_gv["participant_age_group"] = new_gv["participant_age_group"].astype(str)
new_gv["participant_gender"] = new_gv["participant_gender"].astype(str)
new_gv["participant_status"] = new_gv["participant_status"].astype(str)
new_gv["participant_type"] = new_gv["participant_type"].astype(str)
```

```python
test=new_gv["participant_age_group"].str.split('\|\|', expand=False)
test
```

```python
new_gv.head()
```

```python
# test=new_gv['age_group'].map()

new_gv = new_gv.assign(age_group_Adult = new_gv['participant_age_group'].str.count("Adult 18+"))
new_gv = new_gv.assign(age_group_Teen = new_gv['participant_age_group'].str.count("Teen 12-17"))
new_gv = new_gv.assign(age_group_Child = new_gv['participant_age_group'].str.count("Child 0-11"))
new_gv = new_gv.assign(Female = new_gv['participant_gender'].str.count("Female"))
new_gv = new_gv.assign(Male = new_gv['participant_gender'].str.count("Male"))
new_gv = new_gv.assign(Arrested= new_gv['participant_status'].str.count("Arrested"))
new_gv = new_gv.assign(Unharmed = new_gv['participant_status'].str.count("Unharmed"))
new_gv = new_gv.assign(Killed= new_gv['participant_status'].str.count("Killed"))
new_gv = new_gv.assign(Injured= new_gv['participant_status'].str.count("Injured"))
new_gv.head()
```

```python
fixed_gv = new_gv[["incident_id", "date", "state", "city_or_county", "address","incident_url","participant_age","participant_age_group", "participant_gender","participant_status", "participant_type","state_house_district","state_senate_district","age_group_Adult","age_group_Teen","age_group_Child","Female","Male","Arrested","Unharmed","Killed","Injured"]]
fixed_gv.head()
```
